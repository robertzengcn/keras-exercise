```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
df=pd.read_csv("insurance_data.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2,random_state=25)
```

```python
len(X_train)
```

```python
# scale can help model do better
X_train_scaled=X_train.copy()
X_train_scaled['age']=X_train_scaled['age']/100

X_test_scaled=X_test.copy()
X_test_scaled['age']=X_test_scaled['age']/100
```

```python
X_test_scaled
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy']
              )
model.fit(X_train_scaled,y_train,epochs=1000)
```

```python
model.evaluate(X_test_scaled,y_test)
```

```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
sigmoid(18)
```

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon)for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon)for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_decent(age, affordability, y_true, epochs,loss_thresold):
    # try to find w1, w2, bias   
    w1=w2=1
    bias=0 
    rate=0.5
    n=len(age)
    
    for i in range(epochs):
        weighted_sum=w1*age+w2*affordability+bias
        y_predicted=sigmoid_numpy(weighted_sum)
        
        loss=log_loss(y_true,y_predicted)
        
        w1d=(1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d=(1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))
        bias_d=np.mean(y_predicted-y_true)
        
        w1=w1-rate*w1d
        w2=w2-rate*w2d
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        if loss<=loss_thresold:
            break
    
    return w1,w2,bias    
        
```

```python
class myNN:
    def __init__(self):
        self.w1=1
        self.w2=1
        self.bias=0
    def fit(self,X,y,epochs,loss_thresold):
        self.w1,self.w2,self.bias=self.gradient_decent(X['age'],X['affordibility'],y,epochs,loss_thresold)
    def predict(self,X_test):     
        weighted_sum=self.w1*X_test['age']+self.w2*X_test['affordibility']+self.bias
        return sigmoid_numpy(weighted_sum)
    
    def gradient_decent(self,age, affordability, y_true, epochs,loss_thresold):
    # try to find w1, w2, bias   
        w1=w2=1
        bias=0 
        rate=0.5
        n=len(age)
    
        for i in range(epochs):
            weighted_sum=w1*age+w2*affordability+bias
            y_predicted=sigmoid_numpy(weighted_sum)

            loss=log_loss(y_true,y_predicted)

            w1d=(1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
            w2d=(1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))
            bias_d=np.mean(y_predicted-y_true)

            w1=w1-rate*w1d
            w2=w2-rate*w2d
            if i%50==0:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
            if loss<=loss_thresold:
                break
    
        return w1,w2,bias      
```
